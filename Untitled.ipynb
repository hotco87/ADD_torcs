```python
def compare_cluster_result(model, orig_state, aug_state, has_actions=False):
    orig_label = model.predict(orig_state)

    ## TODO : fix it
    aug_label = model.predict(aug_state)
    # aug_label = model.predict(np.float64(aug_state))

    print('{:^5} {:^8} {:^8} {:^6} {:^8}'.format('idx', 'origin', 'augment', 'fluc.', 'matches'))

    for idx in range(model.get_params()['n_clusters']):        
        orig_idx = np.argwhere(orig_label == idx).reshape(-1)
        aug_idx = np.argwhere(aug_label == idx).reshape(-1)
        fluctuation = len(aug_idx) - len(orig_idx)
        matches = set(orig_idx) & set(aug_idx)

        actions = None
        if has_actions:
            actions = set(orig_state[orig_idx][:, -1]) & set(aug_state[aug_idx][:, -1])

        print('{:>4} {:>8} {:>8} {:^ 6} {:>8} {:<32s}'.format(idx, len(orig_idx), len(aug_idx), fluctuation, len(matches), str(actions)))
```

```python
from sklearn import cluster

from scipy.spatial import distance
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
import numpy as np

def compute_bic(kmeans,X):
    """
    Computes the BIC metric for a given clusters

    Parameters:
    -----------------------------------------
    kmeans:  List of clustering object from scikit learn

    X     :  multidimension np array of data points

    Returns:
    -----------------------------------------
    BIC value
    """
    # assign centers and labels
    centers = [kmeans.cluster_centers_]
    labels  = kmeans.labels_
    #number of clusters
    m = kmeans.n_clusters
    # size of the clusters
    n = np.bincount(labels)
    #size of data set
    N, d = X.shape

    #compute variance for all clusters beforehand
    #fixed
    cl_var = (1.0 / (N - m) / d) * sum([sum(distance.cdist(X.iloc[labels == i], [centers[0][i]], 
             'euclidean')**2) for i in range(m)])

    const_term = 0.5 * m * np.log(N) * (d+1)

    BIC = np.sum([n[i] * np.log(n[i]) -
               n[i] * np.log(N) -
             ((n[i] * d) / 2) * np.log(2*np.pi*cl_var) -
             ((n[i] - 1) * d/ 2) for i in range(m)]) - const_term

    return(BIC)


ks = range(1,10)

# run 9 times kmeans and save each result in the KMeans object
KMeans = [cluster.KMeans(n_clusters = i, init="k-means++").fit(X) for i in ks]

# now run for each cluster the BIC computation
BIC = [compute_bic(kmeansi,X) for kmeansi in KMeans]

print BIC
```
